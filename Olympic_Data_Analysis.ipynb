```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# Load dataset
athletes = pd.read_csv("athlete_events.csv")
regions = pd.read_csv("noc_regions.csv")
```

```python
athletes.head()
```

```python
regions.head()
```

```python
# Join the dataframes
athletes_df = athletes.merge(regions, how='left', on='NOC')
athletes_df.head()
```

```python
athletes_df.shape
```

```python
athletes_df.rename(columns={'region':'Region', 'notes':'Notes'}, inplace=True);
```

```python
athletes_df.head()
```

```python
athletes_df.info()
```

```python
athletes_df.describe()
```

```python
nan_values = athletes_df.isnull()
nan_columns = nan_values.any()
nan_columns
```

```python
athletes_df.isnull().sum()
```

```python
# India Details
athletes_df.query('Team == "India"').head(5)
```

```python
# Japan Details
athletes_df.query('Team == "Japan"').head()
```

```python
top_10_countries = athletes_df.Team.value_counts().sort_values(ascending=False).head(10)
top_10_countries
```

```python
# Plot for Top 10 Countries
plt.figure(figsize=(12, 6))
plt.title('Overall Participation by Country')
sns.barplot(x=top_10_countries.index, y=top_10_countries, palette = 'Set2')
```

```python
# Age Distribution of the participants

plt.figure(figsize=(12, 6))
plt.title("Age Distribution of The Participants")
plt.xlabel('Age')
plt.ylabel('Number of participants')
plt.hist(athletes_df.Age, bins = np.arange(10, 80, 2), color='lightblue', edgecolor='white');
```

```python
winter_sports = athletes_df[athletes_df.Season == 'Winter'].Sport.unique()
winter_sports
```

```python
summer_sports = athletes_df[athletes_df.Season == 'Summer'].Sport.unique()
summer_sports
```

```python
gender_counts = athletes_df.Sex.value_counts()
gender_counts
```

```python
# Gender Distribution
plt.figure(figsize=(12, 6))
plt.title('Gender Distribution')
plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%', startangle=180, shadow=True)
```

```python
# Total Medals
athletes_df.Medal.value_counts()
```

```python
female_participants = athletes_df[(athletes_df.Sex=='F') & (athletes_df.Season=='Summer')][['Sex', 'Year']]
female_participants = female_participants.groupby('Year').count().reset_index()
female_participants.tail()
```

```python
womenOlympics = athletes_df[(athletes_df.Sex=='F') & (athletes_df.Season == 'Summer')]
```

```python
# Women Participation by Year
sns.set(style="darkgrid")
plt.figure(figsize=(20, 10))
sns.countplot(x='Year', data=womenOlympics, palette='Spectral')
plt.title('Women Participantion')
```

```python
part = womenOlympics.groupby('Year')['Sex'].value_counts()
plt.figure(figsize=(20, 10))
part.loc[:, 'F'].plot()
plt.title('Plot of Female Athletes over time')
```

```python
# Gold Medal Participants
goldMedals = athletes_df[(athletes_df.Medal == 'Gold')]
goldMedals.head()
```

```python
# Taking only the values ignoring the Nan Valuse
goldMedals = goldMedals[np.isfinite(goldMedals['Age'])]
```

```python
# Gold beyond 60
goldMedals['ID'][goldMedals['Age'] > 60].count()
```

```python
sporting_event = goldMedals['Sport'][goldMedals['Age'] > 60]
sporting_event
```

```python
# Gold medals from each Country
goldMedals.Region.value_counts().reset_index(name='Medal').head()
```

```python
totalGoldMedals = goldMedals.Region.value_counts().reset_index(name='Medal').head()
g = sns.catplot(x='index', y='Medal', data=totalGoldMedals,
                height=5, kind='bar', palette='rocket')
g.despine(left=True)

plt.title('Gold Medals per Country')
```

```python
# Rio Olympics
max_year = athletes_df.Year.max()
print(max_year)
```

```python
team_names = athletes_df[(athletes_df.Year == max_year) & (athletes_df.Medal == 'Gold')].Team
team_names.value_counts().head()
```

```python
sns.barplot(x=team_names.value_counts().head(20), y=team_names.value_counts().head(20).index)

plt.ylabel(None)
plt.xlabel('Countrywise Medal for the Year 2016')
```

```python
not_null_medals = athletes_df[(athletes_df['Height'].notnull()) & (athletes_df['Weight'].notnull())]
```

```python
plt.figure(figsize = (12, 10))
axis = sns.scatterplot(x = 'Height', y = 'Weight', data = not_null_medals, hue='Sex')
plt.title('Height Vs Weight of Olympic Medalists')
```
